```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import easyocr
```

## Image 1

```python
IMAGE_PATH = 'surf.jpeg'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH)
result
```

#### Recognised as 'SURF', with 98.662082004547119 %

```python
# Drawing the results

top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
# Highleting the text and writing the same

img = cv2.imread(IMAGE_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 0.5,(0,0,0),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

## Image 2

```python
IMAGE_PATH = 'sign.png'
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(IMAGE_PATH)
result
```

```python
# Drawing the results and displaying the text

img = cv2.imread(IMAGE_PATH)
spacer = 100
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,0,0),2,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
